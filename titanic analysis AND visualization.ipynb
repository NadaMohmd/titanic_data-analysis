# fill nulls in age using median

```python
df['Age'] = df['Age'].fillna(df.Age.median())
```

# drop the other columns that have nulls

```python
df.drop(['Cabin','Embarked'],inplace = True ,axis =1)
```

```python
df.isnull().sum()
```

# check num of survived by age

```python
df[df['Survived'] == 0]['Age'].hist(color ='gray' , edgecolor ='black')
plt.title('Pepole was survived ')
plt.xlabel('Range of age')
plt.ylabel('Num of survived')
```

# check num of survived and not survived by sex

```python
count = pd.crosstab(df['Sex'],df['Survived'])
count.plot(kind ='bar',color = ['red','blue'])
plt.title('SURVIVED BASED ON SEX')
plt.legend(['NOT SURVIVED','SURVIVED'])
```

# show count of sex using bar

```python
sex_count =df['Sex'].value_counts()
sex_count.plot(kind ='bar')
plt.title('sex count')
```

# check survived and not by pclass

```python
count = pd.crosstab(df['Pclass'],df['Survived'])
count.plot(kind ='bar',color = ['red','blue'])
plt.title('SURVIVED BASED ON Pclass')
plt.legend(['NOT SURVIVED','SURVIVED'])
```

# pie chart of counts of pclass

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Num of pepole in each class')
```

# Use subplot 1*2  

```python
fig = plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df[df['Survived'] == 0]['Age'], bins=20, color='gray', edgecolor ='black')
plt.title('People who did not survive')
plt.xlabel('Range of Age')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(df[df['Survived'] == 1]['Age'], bins=20, color='blue', edgecolor ='black')
plt.title('People who survived')
plt.xlabel('Range of Age')
plt.ylabel('Count')

plt.tight_layout()  
plt.show()
```
